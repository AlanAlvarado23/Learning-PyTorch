# Introducción

PyTorch es una biblioteca de aprendizaje automático de código abierto desarrollada por el equipo de investigación de inteligencia artificial de Facebook. Es una biblioteca basada en Python que se utiliza para construir redes neuronales profundas y admite gráficos de cálculo dinámico. En esta unidad, trabajaremos:
- Cuáles son las características principales de PyTorch para convertirla en una biblioteca de aprendizaje profundo
- Cómo instalar PyTorch
- Algunos ejemplos básicos de código PyTorch

# PyTorch a un alto nivel
- PyTorch ofrece dos características clave de alto nivel:
  1. Cálculo tensorial (similar a NumPy, con opción de aceleración GPU)
  2. Redes neuronales profundas basadas en diferenciación automática

- Ventajas de PyTorch:
  - Manipulación eficiente de números mediante tensores
  - Aceleración de cálculos con GPU
  - Diferenciación automática para ajuste de parámetros

- PyTorch es popular por:
  - Simplicidad y facilidad de uso
  - Naturaleza dinámica (apreciada en investigación)
  - Facilidad para modificar arquitecturas de modelo en tiempo real

- Características clave:
  - API simple e intuitiva
  - Integración con NumPy
  - Gráficos de cálculo dinámico
  - Soporte para entrenamiento distribuido a través de GPU
  - Comunidad grande y activa

- Resumen:
  - Biblioteca potente y fácil de usar
  - Ampliamente utilizada en academia e industria
  - Excelente opción para iniciarse en aprendizaje profundo
  - Combina naturaleza dinámica, facilidad de uso y fuerte apoyo comunitario

Una vez que haya instalado PyTorch, puede comenzar a experimentar con él. Aquí hay un ejemplo simple que le permitirá verificar que su PyTorch esté funcionando:

```python
import torch as pt

x = pt.tensor([[1, 2, 3], [4, 5, 6]])

y = pt.tensor([[1, 2],[3, 4], [5, 6]])

z = pt.matmul(x, y)

print(x)
print(y)
print(z)
```

# Aplicaciones de PyTorch
PyTorch se utiliza ampliamente en diversas áreas del aprendizaje automático y aprendizaje profundo:
- Visión por computadora:
  - Tareas: clasificación de imágenes, detección de objetos, segmentación
  - Modelos populares: ResNet, YOLO, Faster R-CNN
- Procesamiento del lenguaje natural (NLP):
  - Aplicaciones: análisis de sentimientos, traducción automática, generación de texto
  - Modelos destacados: BERT, Transformer
- Aprendizaje por refuerzo:
  - Usos: juegos, control de robots
  - Modelos disponibles: DQN, PPO, A3C
- Otras aplicaciones:
  - Reconocimiento de voz
  - Previsión de series temporales
  - Detección de anomalías

# Ventajas y limitaciones de PyTorch
- Ventajas de PyTorch:
  - Gráficos de cálculo dinámico:
    - Generados sobre la marcha durante la ejecución
    - Permite codificación de alto nivel
    - Facilita experimentación rápida
  - Facilidad de uso:
    - API simple e intuitiva
    - Rápida creación de prototipos
    - Instalación y configuración sencillas
  - Aceleración con GPU:
    - Soporte para aceleración GPU
    - Mayor velocidad que otras bibliotecas
  - Comunidad activa:
    - Gran comunidad de soporte
    - Desarrollo continuo de la biblioteca
    - Acceso a nuevas funciones y ayuda rápida
- Limitaciones de PyTorch:
  - Uso limitado en producción:
    - Excelente para investigación y prototipo
    - Menos utilizado en entornos de producción
  - Curva de aprendizaje:
    - Puede ser pronunciada para principiantes
    - Desafío para nuevos en aprendizaje profundo

# PyTorch VS. TensorFlow
- PyTorch y TensorFlow son competidores principales en aprendizaje profundo.
- Ventajas de TensorFlow:
  - Interfaz primitiva para operaciones de bajo nivel
  - Integración estrecha con Keras en TensorFlow 2
  - Mejor para producción y despliegue en dispositivos móviles/edge
- Ventajas de PyTorch:
  - Mayor flexibilidad y control para usuarios avanzados
  - Gráficos de cálculo dinámico, sin tensores de marcador de posición
  - Ideal para investigación y creación de prototipos
- Proyectos asociados:
  - PyTorch: `torchvision` para visión por computadora
  - TensorFlow: 
    - TensorFlow.js para navegadores
    - TensorFlow Lite para dispositivos con memoria limitada
    - TensorBoard para visualización (también compatible con PyTorch)
- Resumen:
  - PyTorch: preferido para investigación y prototipos
  - TensorFlow: preferido para producción y despliegue en dispositivos
La elección depende de las necesidades específicas del proyecto y las preferencias del usuario.
