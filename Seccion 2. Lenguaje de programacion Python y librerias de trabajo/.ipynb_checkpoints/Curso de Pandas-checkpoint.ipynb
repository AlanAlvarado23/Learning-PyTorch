## Curso de Pandas

Pandas proporcina estructuras de datos y funcionalidad para manipular y analizar datos rápidamente. La clave para comprender Pandas para el aprendizaje automático es comprender sus estructuras de datos básicas: Series y DataFrame
### Series
Una serie es un array unidimensional donde se pueden etiquetar las filas y columnas

```python
import numpy as np
import pandas as pd

my_array = np.array([1, 2, 3])
row_name = ['a', 'b', 'c']

serie = pd.Series(my_array, index=row_name)
serie
```

Acceder a los datos:

```python
print(serie['a'])
```

### DataFrame
Un DataFrame es un array multidimensional donde se pueden etiquetar las filas y las columnas

```python
import numpy as np
import pandas as pd

my_array = np.array([[1, 2, 3],
                     [4, 5, 6]])

row_names = ['a', 'b']
col_names = ['one', 'two', 'three']

dataframe = pd.DataFrame(my_array, index=row_names, columns=col_names)
dataframe
```

```python
print(f"One column: {dataframe['one']}")
print(f"One column: {dataframe.one}")
```
